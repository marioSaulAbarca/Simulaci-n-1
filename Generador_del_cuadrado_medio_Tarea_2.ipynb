```python
import math
from tabulate import tabulate

# Función para el Generador del Cuadrado Medio
def generador_cuadrado_medio(semilla, n):
    resultados = []
    longitud_semilla = len(str(semilla))

    for i in range(n):
        # Elevar la semilla al cuadrado
        cuadrado = semilla ** 2

        # Convertir a string para extraer los dígitos del medio
        cuadrado_str = str(cuadrado).zfill(2 * longitud_semilla)
        mitad = len(cuadrado_str) // 2
        nuevo_valor = int(cuadrado_str[mitad - longitud_semilla//2: mitad + longitud_semilla//2])

        # Normalizar el valor entre 0 y 1
        numero_normalizado = nuevo_valor / (10 ** longitud_semilla)

        # Agregar a la tabla de resultados
        resultados.append([i+1, semilla, cuadrado, nuevo_valor, numero_normalizado])

        # Actualizar la semilla
        semilla = nuevo_valor

    # Imprimir la tabla ordenada
    print(tabulate(resultados, headers=["Iteración", "Semilla", "Cuadrado", "Nuevo Valor", "Número Normalizado"], tablefmt="grid"))

# Parámetros iniciales
semilla = int(input("Ingresa la semilla (número entero): "))
n = int(input("Ingresa la cantidad de números aleatorios a generar: "))

# Ejecutar el generador
generador_cuadrado_medio(semilla, n)
```
